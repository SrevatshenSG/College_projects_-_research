```python
import cv2
import torch
import time
import os
import numpy as np
import psutil
import pynvml
from collections import defaultdict

# Initialize GPU monitoring
pynvml.nvmlInit()
gpu_handle = pynvml.nvmlDeviceGetHandleByIndex(0)  

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
model.conf = 0.4  # Confidence threshold

# Video settings
video_path = "downloaded_video1.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Directory for sequential logs
log_dir = "sequential_logs"
os.makedirs(log_dir, exist_ok=True)  # Create directory if it doesn't exist

# Logging files
log_file = os.path.join(log_dir, "detection_log.txt")
fps_log = os.path.join(log_dir, "fps_log.txt")
resource_log = os.path.join(log_dir, "resource_log.txt")  # New resource log file

# Clear logs before running
open(log_file, "w").close()
open(fps_log, "w").close()
open(resource_log, "w").close()  # Clear the resource log before running

# Tracking
frame_count = 0
start_time = time.time()
paused = False  # Pause flag

while cap.isOpened():
    if not paused:
        ret, frame = cap.read()
        if not ret:
            break  # Stop when video ends

        frame_count += 1

        # Perform detection
        results = model(frame)

        # Object count per frame
        object_counts = defaultdict(int)
        detections = results.xyxy[0].cpu().numpy()

        for det in detections:
            cls = int(det[-1])  # Class index
            object_counts[cls] += 1  

        # Calculate FPS
        elapsed_time = time.time() - start_time
        fps = frame_count / elapsed_time if elapsed_time > 0 else 0

        # Log object counts
        with open(log_file, "a") as f:
            for cls, count in object_counts.items():
                f.write(f"{cls},{count}\n")

        # Log FPS
        with open(fps_log, "a") as f:
            f.write(f"{fps}\n")

        # Log resource usage (Lightweight approach)
        if frame_count % 5 == 0:  # Log resource data every 5 frames to reduce overhead
            cpu_percent = psutil.cpu_percent(interval=0.1)
            ram_usage = psutil.virtual_memory().used / (1024 ** 2)  # in MB
            gpu_util = pynvml.nvmlDeviceGetUtilizationRates(gpu_handle).gpu  # in percentage
            gpu_mem = pynvml.nvmlDeviceGetMemoryInfo(gpu_handle).used / (1024 ** 2)  # in MB

            # Write resource usage to log file
            with open(resource_log, "a") as f:
                f.write(f"{frame_count},{cpu_percent},{ram_usage},{gpu_util},{gpu_mem}\n")

        # Display detections
        results.render()
        cv2.imshow("YOLOv5 Object Detection", results.ims[0])

        print(f"[LOG] Frame {frame_count}: FPS={fps:.2f}, Objects Detected={sum(object_counts.values())}")

    # Keyboard controls
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break  # Quit
    elif key == ord(' '):
        paused = not paused  # Toggle pause

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import torch
import time
import os
import numpy as np
import psutil
from collections import defaultdict
from concurrent.futures import ThreadPoolExecutor
from pynvml import nvmlInit, nvmlDeviceGetHandleByIndex, nvmlDeviceGetUtilizationRates, nvmlDeviceGetMemoryInfo

# Initialize GPU monitoring
nvmlInit()
handle = nvmlDeviceGetHandleByIndex(0)  

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
model.conf = 0.4  # Confidence threshold

# Video settings
video_path = "downloaded_video1.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Logging directory
log_dir = "threading_logs"
os.makedirs(log_dir, exist_ok=True)
log_file = os.path.join(log_dir, "detection_log.txt")
fps_log = os.path.join(log_dir, "fps_log.txt")
resource_log = os.path.join(log_dir, "resource_log.txt")  #  New file

# Clear logs before running
open(log_file, "w").close()
open(fps_log, "w").close()
open(resource_log, "w").close()  # lear resource log

# Tracking
frame_count = 0
start_time = time.time()
paused = False  # Pause flag

def process_frame(frame):
    """Processes a single frame (object detection)."""
    results = model(frame)

    # Object count per frame
    object_counts = defaultdict(int)
    detections = results.xyxy[0].cpu().numpy()

    for det in detections:
        cls = int(det[-1])  # Class index
        object_counts[cls] += 1  

    return results, object_counts

# Thread pool
executor = ThreadPoolExecutor(max_workers=2)

while cap.isOpened():
    if not paused:
        ret, frame = cap.read()
        if not ret:
            break  # Stop when video ends

        frame_count += 1

        # Process frame in parallel
        future = executor.submit(process_frame, frame)
        results, object_counts = future.result()

        # Calculate FPS
        elapsed_time = time.time() - start_time
        fps = frame_count / elapsed_time if elapsed_time > 0 else 0

        # Log object counts
        with open(log_file, "a") as f:
            for cls, count in object_counts.items():
                f.write(f"{frame_count},{cls},{count}\n")

        # Log FPS
        with open(fps_log, "a") as f:
            f.write(f"{frame_count},{fps:.2f}\n")

        #  Log resource usage
        cpu_percent = psutil.cpu_percent()
        ram_used = psutil.virtual_memory().used / (1024 ** 2)  # Convert bytes to MB
        gpu_util = nvmlDeviceGetUtilizationRates(handle).gpu
        gpu_mem = nvmlDeviceGetMemoryInfo(handle).used / (1024 ** 2)  # MB

        with open(resource_log, "a") as f:
            f.write(f"{frame_count},{cpu_percent:.2f},{ram_used:.2f},{gpu_util:.2f},{gpu_mem:.2f}\n")

        # Display detections
        results.render()
        cv2.imshow("YOLOv5 Object Detection", results.ims[0])

        print(f"[LOG] Frame {frame_count}: FPS={fps:.2f}, Objects Detected={sum(object_counts.values())}")

    # Keyboard controls
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break  # Quit
    elif key == ord(' '):
        paused = not paused  # Toggle pause

cap.release()
cv2.destroyAllWindows()
executor.shutdown()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import os
from collections import defaultdict

# File paths
log_dir = "logs"
log_file = os.path.join(log_dir, "detection_log.txt")
fps_log = os.path.join(log_dir, "fps_log.txt")
resource_log = os.path.join(log_dir, "resource_log.txt")  

# Check if logs exist
if not os.path.exists(log_file) or not os.path.exists(fps_log):
    print("Error: Log files not found. Run the object detection script first.")
    exit()

# Read FPS data
fps_values = []
with open(fps_log, "r") as f:
    for line in f:
        parts = line.strip().split(",")
        if len(parts) == 2:
            try:
                fps_values.append(float(parts[1]))
            except ValueError:
                print(f"Skipping invalid FPS entry: {line.strip()}")

# Read object detection data
frame_timestamps = []
object_counts_per_frame = defaultdict(list)
total_objects_per_frame = []
total_objects_overall = 0

with open(log_file, "r") as f:
    frame_id = 0
    for line in f:
        parts = line.strip().split(",")
        if len(parts) == 3:
            try:
                cls = int(parts[1])
                count = int(parts[2])
                object_counts_per_frame[cls].append(count)
                total_objects_overall += count
            except ValueError:
                print(f"Skipping invalid object count entry: {line.strip()}")

        frame_timestamps.append(frame_id)
        frame_id += 1
        total_objects_per_frame.append(total_objects_overall)

# Calculate mean values
mean_fps = np.mean(fps_values) if fps_values else 0
mean_objects_per_frame = total_objects_overall / len(frame_timestamps) if frame_timestamps else 0

# Plot 1: FPS Over Time
plt.figure(figsize=(10, 5))
plt.plot(range(len(fps_values)), fps_values, marker='o', linestyle='-', color='b')
plt.axhline(mean_fps, color='r', linestyle='dashed', label=f"Mean FPS: {mean_fps:.2f}")
plt.xlabel("Frame Number")
plt.ylabel("Frames Per Second (FPS)")
plt.title("FPS Over Time")
plt.legend()
plt.grid()
plt.show()

# Plot 2: Object Count Per Frame (Stacked Bar)
plt.figure(figsize=(10, 5))
for cls, counts in object_counts_per_frame.items():
    plt.bar(range(len(counts)), counts, label=f"Class {cls}")
plt.xlabel("Frame Number")
plt.ylabel("Object Count")
plt.title("Object Detection Count Per Frame")
plt.legend()
plt.grid()
plt.show()

# Plot 3: Total Objects Over Time
plt.figure(figsize=(10, 5))
plt.plot(range(len(total_objects_per_frame)), total_objects_per_frame, marker='o', linestyle='-', color='g')
plt.xlabel("Frame Number")
plt.ylabel("Cumulative Object Count")
plt.title("Total Objects Detected Over Time")
plt.grid()
plt.show()

# Plot 4: Object Distribution Pie Chart
plt.figure(figsize=(7, 7))
total_per_type = {cls: sum(counts) for cls, counts in object_counts_per_frame.items()}
plt.pie(total_per_type.values(), labels=[f"Class {cls}" for cls in total_per_type.keys()], autopct='%1.1f%%')
plt.title("Total Object Distribution")
plt.show()

# Summary
print("\n🔹 **Metrics Summary:**")
print(f"--> Mean FPS: {mean_fps:.2f}")
print(f"--> Mean Object Count per Frame: {mean_objects_per_frame:.2f}")
print(f"--> Total Objects Detected Across All Frames: {total_objects_overall}")

# Additional Section: System Resource Usage Visualization
resource_log = os.path.join(log_dir, "resource_log.txt")

if os.path.exists(resource_log):
    resource_data = {
        "frame_id": [],
        "cpu_percent": [],
        "ram_usage_mb": [],
        "gpu_util_percent": [],
        "gpu_mem_mb": []
    }

    with open(resource_log, "r") as f:
        for line in f:
            try:
                parts = line.strip().split(",")
                if len(parts) == 5:
                    frame_id = int(parts[0])
                    cpu = float(parts[1])
                    ram = float(parts[2])
                    gpu = float(parts[3])
                    gpu_mem = float(parts[4])

                    resource_data["frame_id"].append(frame_id)
                    resource_data["cpu_percent"].append(cpu)
                    resource_data["ram_usage_mb"].append(ram)
                    resource_data["gpu_util_percent"].append(gpu)
                    resource_data["gpu_mem_mb"].append(gpu_mem)
            except ValueError:
                print(f"[WARN] Skipping malformed resource_log entry: {line.strip()}")

    # 📊 Plot CPU, RAM, GPU over time
    fig, axs = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle("System Resource Usage Over Time", fontsize=16)

    axs[0, 0].plot(resource_data["frame_id"], resource_data["cpu_percent"], color='orange')
    axs[0, 0].set_title("CPU Usage (%)")
    axs[0, 0].set_xlabel("Frame")
    axs[0, 0].set_ylabel("CPU %")
    axs[0, 0].grid()

    axs[0, 1].plot(resource_data["frame_id"], resource_data["ram_usage_mb"], color='purple')
    axs[0, 1].set_title("RAM Usage (MB)")
    axs[0, 1].set_xlabel("Frame")
    axs[0, 1].set_ylabel("RAM (MB)")
    axs[0, 1].grid()

    axs[1, 0].plot(resource_data["frame_id"], resource_data["gpu_util_percent"], color='green')
    axs[1, 0].set_title("GPU Utilization (%)")
    axs[1, 0].set_xlabel("Frame")
    axs[1, 0].set_ylabel("GPU %")
    axs[1, 0].grid()

    axs[1, 1].plot(resource_data["frame_id"], resource_data["gpu_mem_mb"], color='red')
    axs[1, 1].set_title("GPU Memory Usage (MB)")
    axs[1, 1].set_xlabel("Frame")
    axs[1, 1].set_ylabel("GPU Mem (MB)")
    axs[1, 1].grid()

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
else:
    print("[INFO] Skipping resource graph: resource_log.txt not found.")

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Replace these arrays with your actual runtime data (in seconds) for each method
runtime_seq = np.array([10, 9, 8, 7, 6])      # Sequential or baseline
runtime_thread = np.array([8, 7, 6, 5, 4])    # Threading or method 2
runtime_parallel = np.array([6, 5, 4, 3, 2])  # Multiprocessing/parallel or method 3

# Calculate speedup (baseline / method)
speedup_thread = runtime_seq / runtime_thread
speedup_parallel = runtime_seq / runtime_parallel

# X-axis values (e.g., test cases, input sizes, number of cores, etc.)
x = np.arange(1, len(runtime_seq) + 1)

plt.figure(figsize=(12, 6))

# --- Runtime Comparison ---
plt.subplot(1, 2, 1)
plt.plot(x, runtime_seq, marker='o', label='Sequential')
plt.plot(x, runtime_thread, marker='o', label='Threading')
plt.plot(x, runtime_parallel, marker='o', label='Parallel')
plt.xlabel('Test Case')
plt.ylabel('Runtime (seconds)')
plt.title('Runtime Comparison')
plt.legend()
plt.grid(True)

# --- Speedup Comparison ---
plt.subplot(1, 2, 2)
plt.plot(x, np.ones_like(x), '--', label='Sequential (baseline)')
plt.plot(x, speedup_thread, marker='o', label='Threading Speedup')
plt.plot(x, speedup_parallel, marker='o', label='Parallel Speedup')
plt.xlabel('Test Case')
plt.ylabel('Speedup (vs Sequential)')
plt.title('Speedup Comparison')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
